2.   언어
>**구글 번역기**
 : 구글 번역은 전 세계에서 가장 널리 이용되는 언어 번역 서비스로 '*신경망 번역* ' 즉, 머신러닝을 활용한 번역 서비스이다.
 >
 >* 기존 번역기 (=어구 기반 기계 번역) : 단어 또는 어구를 먼저 번역한 후 이를 퍼즐 맞추듯이 조립해야 했기 때문에 문장의 구조가 매우 어색하고, 문장 속에 담긴 작성자의 참뜻을 이해하지 못한다.
 >
 >* 방법 : 머신러닝을 통해 문장 전체의 흐름을 인식하고 문장 속에 담긴 작성자의 의도를 이해해서 번역을 진행하는 프로그래밍이다. 즉, 사람처럼 문장을 통째로 이해한 후 번역하는 것이다.
 >* 개발 순서 : 인터넷에 돌아다니는 인간이 미리 번역한 콘텐츠를 웹 크롤링을 통해 수집하고, 번역 초기 시스템을 구축한 뒤 일일이 개발자의 감독하에 지도 학습을 받으면서 이를 바탕으로 스스로 성능이 강화되는 순서로 개발되었다.
>
>* 결과 : 신경망 번역을 도입함 (자연스럽고 더욱 정확한 번역 서비스를 제공 가능) 번역 오류의 감소 ( 55~80%까지 줄어들었다.)
>
>* 특징 : 번역 모델을 구축할 때 언어별로 모델을 각각 만들지 않고, 구조가 비슷한 언어끼리 묶어서 모델을 구축. EX) 한국어의 경우 언어 구조가 비슷한 일본어, 터키어와 번역 모델을 공유한다.
>
>* 장점 : 
>>1. 유사한 언어끼리 번역 모델을 공유하면 번역 모델의 수를 크게 줄일 수 있고, 각각의 번역 모델이 서로 학습한 내용을 공유할 수도 있기 때문에 번역 모델 구축에 들어가는 리소스를 크게 줄일 수 있다.
>>
>>2. 하나의 언어에 대한 학습이 유사한 언어까지 같이 학습되기에 사용자가 적은 언어도 정확하고 매끄러운 번역이 가능해진다.

4. 자율주행
>**SCC-ML** 
 :　현대자동차와 기아자동차에서 개발한 운전자의 주행성향에 맞는 ‘부분 자율주행’을 구현하는 기술이다.
>
>* 방법 : SCC기술에 AI를 추가하여 운전자의 주행 성향을 차가 자가학습해 운전자와 흡사한 패턴으로 자율주행을 시행한다.
>
>* 원리 : 먼저 전방카메라, 레이더 등의 다양한 센서가 여러 운전상황에서 생기는 정보를 연속적으로 수집해 ADAS의 두뇌격인 제어 컴퓨터로 보낸다. 제어 컴퓨터는 입력된 정보로부터 운전자의 주행습관을 추출하여 종합적인 주행성향을 파악한다.(머신러닝 사용)
>
>* SCC(스마트 크루즈 컨트롤) : ADAS(첨단 운전자보조 시스템)의 주요 기술 중 하나로 앞차와의 거리를 일정하게 유지해 운전자가 설정한 속도로 자율주행을 해주는 기능이다.
>>
>>* 특징(SCC) : 앞차와의 거리, 가속성 등의 주행패턴을 운전자가 직접 설정해야 했으며 조절되는 단계가 세밀하지 않아 운전성향을 고스란히 반영할 수 없었다. 동일한 운전자라 하더라도 가속성향이 고속과 중속, 저속 구간에서 각각 다르지만 SCC는 이런 세부적인 설정을 변경할 수 없었기 때문에 운전자는 이질감을 느끼거나 심할 경우 불안감 때문에 SCC 사용을 꺼리는 경우도 더러 있었다.
>* 특징(SCC-ML) : 주행성향은 크게 보면 ‘앞차와의 거리’, ‘얼마나 신속하게 가속하는지’, ‘주행환경에 얼마나 민첩하게 반응하는지’로 나눌 수 있다. 이런 주행성향과 더불어 다양한 속도, 주변 차량과의 거리 조건조차 모두 고려한 기술이다. 또한, 주행성향에 대한 정보는 센서를 통해 계속 업데이트되기 때문에 운전자의 최근 성향을 반영할 수 있으며 안전운전을 크게 벗어난 주행성향은 따르지 않도록 설정돼 있어서 사고 가능성을 줄이고자 하였다.
